**pytutor-widget** is an animated, [`turtle`](https://docs.python.org/3/library/turtle.html)-compatible
drawing widget for Jupyter notebooks. You write ordinary turtle code in Python; the widget
records each command and replays it as a smooth canvas animation right below the cell.

Built on [`anywidget`](https://anywidget.dev), it behaves identically in **VS Code**,
**JupyterLab**, **Notebook 7**, and **Colab**, and works offline.

## Installation

::: {.panel-tabset}

## pixi

```bash
pixi workspace channel add munch-group
pixi add pytutor-widget
```

## conda

```bash
conda install -c munch-group pytutor-widget
```

## pip

```bash
pip install pytutor-widget
```

:::

````python
from pytutor_widget import Turtle

t = Turtle()
t.speed(6)
for _ in range(4):     # a square
    t.forward(120)
    t.left(90)
````
