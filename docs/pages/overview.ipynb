**codelens-widget** brings the [CodeLens Tool](https://runestone.academy/ns/books/published/overview/Visualizers/codelens.html) used [Python Tutor](https://pythontutor.com/) to the Jupyter Notebook as a widget triggered by a `%%codelens` cell magic.

Built on [`anywidget`](https://anywidget.dev), it behaves identically in **VS Code**, **JupyterLab**, **Notebook 7**, and **Colab**, and works offline.

## Installation

::: {.panel-tabset}

## pixi

```bash
pixi workspace channel add munch-group
pixi add codelens-widget
```

## conda

```bash
conda install -c munch-group codelens-widget
```

## pip

```bash
pip install codelens-widget
```

:::

````python
import codelens_widget
````

````python
%%codelens
def insertion_sort(a):
    for i in range(1, len(a)):
        key = a[i]
        j = i - 1
        while j >= 0 and a[j] > key:
            a[j + 1] = a[j]
            j -= 1
        a[j + 1] = key
    return a

data = [5, 2, 9, 1]
insertion_sort(data)
````
